```python
from langchain.chat_models import ChatOpenAI
from langchain.prompts import PromptTemplate
from langchain.prompts.few_shot import FewShotPromptTemplate
from langchain.callbacks import StreamingStdOutCallbackHandler

chat = ChatOpenAI(
    temperature=0.7,
    model="gpt-3.5-turbo-0125",
    streaming=True,
    callbacks=[
        StreamingStdOutCallbackHandler()
    ]
)

examples = [
    {
        "question" : "catchmind 게임에서 사용할 수 있는 '동물' 주제 5개를 제시해. 대답은 모두 단어여야만 해",
        "answer" : """
            [고양이, 강아지, 참새, 호랑이, 코끼리]
        """
    },
    {
        "question" : "catchmind 게임에서 사용할 수 있는 '하늘' 주제 5개를 제시해 대답은 모두 단어여야만 해",
        "answer" : """
            [독수리, 비행기, 구름, 풍선, 열기구]
        """
    },
    {
        "question" : "catchmind 게임에서 사용할 수 있는 '바다' 주제 5개를 제시해 대답은 모두 단어여야만 해",
        "answer" : """
            [배, 등대, 소라, 돌고래, 상어]
        """
    }
]

example_template = """
    "Human" : {question}
    "AI" : {answer}
"""

example_prompt = PromptTemplate.from_template(
    "Human : {question} \n AI : {answer}"
)

prompt = FewShotPromptTemplate(
    example_prompt=example_prompt,
    examples=examples,
    suffix="""
    catchmind 게임은 총 4명이서 하는 게임이고, 서로 돌아가면서 특정 주제에 대해서 그림을 그리고 다른 사람은 그걸 맞추는 게임이야.
    catchmind 게임에서 사용할 수 있는 {subject} 주제들에 대해 각각 5개씩 제시해. 
    단 아래 조건을 반드시 따라야해. 
    1. 중복된 단어는 없어야하고 json 형식으로 제시해
    2. catchmind 게임에 맞는 주제를 제시해
    3. 이름 주제를 제외한 나머지는 주제는 반드시 한 단어로 되어있어야해
    """,
    input_variables=["subject"]
)

chain = prompt | chat
chain.invoke({
    "subject" : ["컴퓨터 게임", "인물", "직업", "동물", "일본 애니메이션"],
})

```

```python
from langchain.chat_models import ChatOpenAI
from langchain.prompts import PromptTemplate
from langchain.prompts.few_shot import FewShotPromptTemplate
from langchain.callbacks import StreamingStdOutCallbackHandler

chat = ChatOpenAI(
    temperature=0.1,
    model="gpt-3.5-turbo-0125",
    streaming=True,
    callbacks=[
        StreamingStdOutCallbackHandler()
    ]
)

examples = [
    {
        "question" : "Skribbl.io 게임에서 사용할 수 있는 대주제 5개를 제시해",
        "answer" : """
            [동물, 게임, 인물, 바다, 하늘]
        """
    }
]

example_template = """
    "Human" : {question}
    "AI" : {answer}
"""

example_prompt = PromptTemplate.from_template(
    "Human : {question} \n AI : {answer}"
)

prompt = FewShotPromptTemplate(
    example_prompt=example_prompt,
    examples=examples,
    suffix="Skribbl.io 게임에서 사용할 수 있는 대주제 {number}개를 제시해",
    input_variables=["subject"]
)

chain = prompt | chat
chain.invoke({
    "number" : "3"
})

```
